## After Removing Outliers

```python
for cols in delhi_air.columns[:-1]:
    if delhi_air[cols].dtype == object:
        continue
    else:
        Q1 = delhi_air[cols].quantile(q=0.25)
        Q3 = delhi_air[cols].quantile(q=0.75)
        iqr = Q3 - Q1
    
        low = Q1 - 1.5*(Q3-Q1)
        high = Q3 + 1.5*(Q3-Q1)
        delhi_air.loc[ (delhi_air[cols]>high)] = delhi_air[cols].median()
```

```python
plt.figure(figsize=(15, 6))
sns.set_style("white")
plt.subplot(141)
sns.lineplot(data=delhi_air, x="co", y="pm10")

plt.subplot(142)
sns.lineplot(data=delhi_air, x="pm2_5", y="pm10")

plt.subplot(143)
sns.lineplot(data=delhi_air, x="no", y="pm10")

plt.subplot(144)
sns.lineplot(data=delhi_air, x="no2", y="pm10")
plt.show()
```

```python
# removing outliers
x = delhi_air.drop(['date','pm10'], axis=1)
y = delhi_air.pm10
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.20, 
                                                   random_state=0)
```

```python
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression

car_model = LinearRegression()
car_model.fit(x_train, y_train)
```

```python
m = car_model.coef_
c = car_model.intercept_

m,c
```

```python
y_pred_train = car_model.predict(x_train)
y_pred_train
```

```python
y_pred_test = car_model.predict(x_test)
y_pred_test
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mean_squared_error(y_test, y_pred_test)
```

```python
accuracy = r2_score(y_test, y_pred_test)
print("the model has an accuracy of", accuracy*100)
```

```python
r2score_train= r2_score(y_train, y_pred_train)
r2score_train
```

```python
r2score_test= r2_score(y_test, y_pred_test)
r2score_test
```

```python
car_df = pd.DataFrame({"Actual": y_test,
                       "Predicted": y_pred_test,
                       "Error": abs(y_test-y_pred_test)})

car_df.head().style.background_gradient(cmap="RdYlGn")
```
